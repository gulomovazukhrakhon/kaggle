```python
import torch
import numpy as np
import cv2 as cv

import matplotlib.pyplot as plt
```

```python
# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
model.classes = [2] # cars only
model.conf = 0.6 # confidence score more than 60%

# Image
img = '/kaggle/input/vehicles/vehicles/imgs/seattle12.jpg'

# Inference
results = model(img)

results.show()
```
